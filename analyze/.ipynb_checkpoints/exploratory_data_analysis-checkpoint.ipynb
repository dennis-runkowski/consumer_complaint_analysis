```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
consumer_df = pd.read_csv('../data/complaints.csv')
consumer_df.fillna('',inplace=True)
```

```python
# Format dates and calculate the number of days to respond
def respond(x):
    d = x['Date sent to company'] - x['Date received']
    return d.days
consumer_df['Date received'] =  pd.to_datetime(consumer_df['Date received'], format='%Y-%m-%d')
consumer_df['Date sent to company'] =  pd.to_datetime(consumer_df['Date sent to company'], format='%Y-%m-%d')
consumer_df['Days to Respond'] = consumer_df.apply(respond, axis = 1)
```

```python
# Average days to respond for the full population
consumer_df['Days to Respond'].mean()
```

```python
# Compare the Top 5 US banks
banks = consumer_df[
    (consumer_df.Company == 'U.S. BANCORP') | 
    (consumer_df.Company == 'CITIBANK, N.A.') | 
    (consumer_df.Company == 'BANK OF AMERICA, NATIONAL ASSOCIATION') | 
    (consumer_df.Company == 'WELLS FARGO & COMPANY') | 
    (consumer_df.Company == 'CAPITAL ONE FINANCIAL CORPORATION')
]

sns.set(rc={"figure.figsize":(15, 8)})
chart = sns.countplot(x=banks.Company, data=banks)
for p in chart.patches:
    chart.annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                 textcoords='offset points')
plt.xticks(rotation=45)
```

```python
# Create subset of the data just for the past 3 years of BOA
sub_df = consumer_df[consumer_df.Company == 'BANK OF AMERICA, NATIONAL ASSOCIATION']


indexNames = sub_df[sub_df['Date received'] < '2019-01-01'].index
sub_df.drop(indexNames , inplace=True)


indexNames = sub_df[sub_df['Date received'] >= '2022-01-01'].index
sub_df.drop(indexNames , inplace=True)

# Method to get the year 
def get_year(x):
    year = x.year
    return year

# Tag retail banking products
def retail(x):
    if x in ['Checking or savings account', 'Credit card or prepaid card', 'Payday loan, title loan, or personal loan']:
        return 'Retail'
    return 'Not Retail'
sub_df['Retail'] = sub_df['Product'].apply(retail)


sub_df['received_year'] = sub_df['Date received'].apply(get_year)


sub_df.head()
```

```python
sub_df.info()
```

```python
# Number of complaints for the past three years
sns.set(rc={"figure.figsize":(8, 8)})
sns.set(font_scale = 2)
sns.countplot(x=sub_df.received_year, data=sub_df).set(title='Number of Complaints by Year')
```

```python
# Number of complaints for the past three years - Retail
sns.set(rc={"figure.figsize":(10, 8)})
sns.set(font_scale = 2)
sns.countplot(x=sub_df.received_year, data=sub_df, hue=sub_df.Retail).set(title='Number of Complaints by Year')
```

```python
# Number of complaints by State
sns.set(rc={"figure.figsize":(8, 30)})
sns.set(font_scale = 1)
sns.countplot(y=sub_df.State, data=sub_df).set(title='Number of Complaints by State')
```

```python
# Number of complaints for the past three years by Product
sns.set(rc={"figure.figsize":(16, 8)})
sns.set(font_scale = 2)
sns.countplot(x=sub_df.received_year, data=sub_df, hue=sub_df.Product).set(title='Number of Complaints by Year')
plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
```

```python
# Timely response?
sns.set(rc={"figure.figsize":(15, 30)})
sns.set(font_scale = 2)
sns.countplot(y=sub_df.Issue, data=sub_df[sub_df.Product == 'Checking or savings account']).set(title='Checking or savings account')
```

```python
# Submitted
sns.set(rc={"figure.figsize":(15, 8)})
sns.set(font_scale = 2)
sns.countplot(x=sub_df['Submitted via'], data=sub_df).set(title='Submitted Via')
```

```python
# Product breakdown
sns.set(rc={"figure.figsize":(15, 8)})
sns.set(font_scale = 2)
sns.countplot(y=sub_df.Product, data=sub_df).set(title='Products')
```

```python
# Timely response?
sns.set(rc={"figure.figsize":(15, 8)})
sns.set(font_scale = 2)
sns.countplot(x=sub_df['Timely response?'], data=sub_df).set(title='Timely Response')
```

```python
# Timely response?
sns.set(rc={"figure.figsize":(15, 8)})
sns.set(font_scale = 2)
sns.countplot(y=sub_df['Timely response?'], data=sub_df, hue=sub_df.Product).set(title='Timely Response')
plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
```

```python
# Timely response?
sns.set(rc={"figure.figsize":(15, 30)})
sns.set(font_scale = 2)

large_to_small = sub_df.groupby('Issue').size().sort_values().index[::-1]
sns.countplot(y=sub_df.Issue, data=sub_df, hue=sub_df['Timely response?'],order=large_to_small).set(title='Timely Response')
plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
```
